```python
from scipy import stats
from scipy.optimize import curve_fit
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#path='/content/sample_data/carga.txt'
path='/content/sample_data/cargas.txt'
df = pd.read_csv(path, encoding="utf-8", delimiter="\t")

#cargas
Q = df.carga.values
Q = Q[~np.isnan(Q)] # Para eliminar los NaN (si es que los hay)
#altura
DQ = df.error.values
DQ = DQ[~np.isnan(DQ)]
```

```python
#lista de las diferencias de cargas
diffQ = np.zeros(5671)
#diffQ = np.zeros(666)
alphaQ = np.zeros(5671)
#alphaQ = np.zeros(666)
```

```python
c = 0
for i in range(1,107):
#for i in range(1,37):
  for j in range(i):
    diffQ[c] = abs(Q[i]-Q[j])
    alphaQ[c] = np.sqrt(DQ[i]**2 + DQ[j]**2)
    c = c+1
```

```python
diffQ = diffQ * 10**19
alphaQ = alphaQ * 10**19
```

```python
print(diffQ)
```

```python
print(alphaQ)
```

```python
plt.hist(diffQ, bins=1500, weights=alphaQ)

plt.xlabel('diferencia de cargas')
plt.ylabel('frecuencia')
plt.xlim(0,10)
plt.show()
```
